```python
import math

def quadratic_equation(a, b, c):
    A = b**2 - 4*a*c
    if A > 0:
        x1 = (-b + math.sqrt(A)) / (2*a)
        x2 = (-b - math.sqrt(A)) / (2*a)
        return x1, x2
    elif A == 0:
        x = -b / (2*a)
        return x, x
    else:
        return "No real roots"
```

```python
def test_quadratic_equation():
    assert quadratic_equation(2, 6, 4) == (-1, -2)
    assert quadratic_equation(1, 2, 1) == (-1, -1)
    assert quadratic_equation(4, 6, 3) == "No real roots"
    print("All tests passed")

test_quadratic_equation()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_quadratic(a, b, c):
    x = np.linspace(-10, 10, 400)
    y = a*x**2 + b*x + c

    roots = quadratic_equation(a, b, c)
    if roots != "No real roots":
        plt.plot(roots[0], 0, 'ro')  # plot the first root
        plt.plot(roots[1], 0, 'ro')  # plot the second root

    plt.plot(x, y)
    plt.title("Quadratic function: {}x^2 + {}x + {}".format(a, b, c))
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.show()

plot_quadratic(2, 6, 4)
plot_quadratic(1, 2, 1)
plot_quadratic(4, 6, 3)
```
